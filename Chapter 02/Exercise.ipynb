```python
def add(*vectors):
    return (sum([v[0] for v in vectors]), sum([v[1] for v in vectors]))
```

```python
add(*[u, v])
```

```python
def translate(translation, *vectors):
    return [(translation[0] + v[0], translation[1] + v[1]) for v in vectors]
```

```python
translate((1, 1), *[u, v])
```

```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]
```

```python
draw(
    Points(*dino_vectors),
    Polygon(*dino_vectors)
)
```

```python
def draw_100_dinasours():
    translations = [(x*12, y*10)
                    for x in range(-5, 5)
                    for y in range(-5, 5)]
    dinos = [Polygon(*translate(t, *dino_vectors)) for t in translations]
    draw(*dinos, grid=None, origin=None, axes=None)
```

```python
draw_100_dinasours()
```

```python
z = (-1, 1)
v = (1, 1)
points = [(r * z[0] + s * v[0], r * z[1] + s * v[1])
          for r in range(-2, 3)
          for s in range(0, 1)]
draw(Points(*points))
```

```python
from random import uniform
def randomR():
    return uniform(-3, 3)
def randomS():
    return uniform(-1, 1)
ps = [add(scale(randomR(), z), scale(randomS(), v)) 
      for i in range(0, 500)]
draw(Points(*ps))
```

```python
def perimeter(*vectors):
    return sum(
            [(distance(vectors[i], vectors[(i+1)%len(vectors)])) 
             for i in range(0, len(vectors))])
```

```python
perimeter(*dino_vectors)
```

```python
def regular_polygon(n):
    cords = [(1, 2*pi*i/n) for i in range(0, n)]
    points = [to_cartesian(c) for c in cords]
    draw(Polygon(*points))
```

```python
regular_polygon(9)
```
